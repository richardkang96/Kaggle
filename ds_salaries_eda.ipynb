```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#https://www.kaggle.com/saurabhshahane/data-science-jobs-salaries
df = pd.read_csv('Data Science Jobs Salaries.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(14.5,5),dpi=200)
sns.boxplot(x='experience_level',y='salary_in_usd',data=df,order=['EN','MI','SE','EX']);
```

Entry Level and Mid Level experienced individuals seem to be very comparable in salaries. However, as you go higher to Senior and Executive experience levels, the salaries visibly increase, as expected.

```python
plt.figure(figsize=(14.5,5),dpi=200)
sns.countplot(x='company_size',data=df,order=['S','M','L']);
```

There seems to be large companies the most in this dataset. Small and medium companies are each about half of the total amount of large companies.

```python
plt.figure(figsize=(14.5,5),dpi=200)
sns.countplot(x='salary_currency',data=df,order=list(df.groupby('salary_currency').size().sort_values().index));
```

USD seems to be the most prevalent salary currency in this dataset with EUR coming in at second with about half of USD's total.

```python
plt.figure(figsize=(14.5,5),dpi=200)
sns.scatterplot(x='experience_level',y='salary_in_usd',data=df);
```

EX seems to be the highest in salary overall, however, has the fewest number of data points across all experience levels.

```python
plt.figure(figsize=(14.5,5),dpi=200)
sns.histplot(data=df['salary_in_usd']);
```

The majority of data points have a salary in USD below 150,000, with the mode being between 50,000-75,000.

```python
plt.figure(figsize=(14.5,5),dpi=200)
sns.barplot(x='job_title',y='salary_in_usd',data=df,order=list(df.groupby('job_title')['salary_in_usd'].mean().sort_values().index))
plt.xticks(rotation=90);
```

Financial Data Analysts seem to have the most salary out of any other Data Science position on average. The line (error bar) for Principal Data Engineer is very large. This means there is a lot of uncertainity with the salaries for that position based on the mean.
